```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense, Dropout
from keras.utils import np_utils
from sklearn.model_selection import KFold
from keras.regularizers import l2
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
y_train = train['Category']
y_train = (y_train + 1) / 2

train.pop('Id')
train.pop('Category')
X_train = (train - train.mean()) / train.std()

answer = test['Id']
test.pop('Id')
X_test = (test - test.mean()) / test.std()
```

```python
model = Sequential([
    Dense(5, input_dim = 300, activation = 'relu'),
    Dense(1, 'sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(X_train[:80000], y_train[:80000], validation_data = (X_train[80000:], y_train[80000:]),
                    epochs = 40, batch_size = 64)
test_loss, test_acc = model.evaluate(X_train[80000:], y_train[80000:])
```

```python
kfold = KFold(5)
accuracy = []
for train, test in kfold.split(X_train, y_train):
    model = Sequential([
        Dense(150, input_dim = 300, activation = 'relu'),
        Dense(1, 'sigmoid')
    ])
    print(type(test), train)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    history = model.fit(X_train.loc[train, :], y_train[train], validation_data = (X_train.loc[test, :], y_train[test]),
                    epochs = 10, batch_size = 64)
    test_loss, test_acc = model.evaluate(X_train.loc[test, :], y_train[test])
    accuracy.append(test_acc)
```
